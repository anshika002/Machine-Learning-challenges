```python
c = cv2.resize(center_image,(100,100))
plt.imshow(c)
#plt.axis('off')
plt.show()
```

```python

tl = cv2.resize(top_left,(200,200))
plt.imshow(tl)
#plt.axis('off')
plt.show()
tr = cv2.resize(top_right,(200,200))
plt.imshow(tr)
#plt.axis('off')
plt.show()
br = cv2.resize(bottom_right,(200,200))
plt.imshow(br)

plt.show()
bl = cv2.resize(bottom_left,(200,200))
plt.imshow(bl)
#plt.axis('off')
plt.show()
```

```python
cv2.copyMakeBorder(tl, 10, 10, 10, 10,borderType = cv2.BORDER_CONSTANT)
cv2.copyMakeBorder(tr, 10, 10, 10, 10,borderType = cv2.BORDER_CONSTANT)
cv2.copyMakeBorder(bl, 10, 10, 10, 10,borderType = cv2.BORDER_CONSTANT)
cv2.copyMakeBorder(br, 10, 10, 10, 10,borderType = cv2.BORDER_CONSTANT)
cv2.copyMakeBorder(c, 10, 10, 10, 10,borderType = cv2.BORDER_CONSTANT)
```

```python
top = np.hstack([tl, tr])
bottom = np.hstack([bl, br])
merge = np.vstack([top, bottom])
merge[100*300 : 100*300,:] = c[]
```

```python
image = cv2.cvtColor(merge,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```
